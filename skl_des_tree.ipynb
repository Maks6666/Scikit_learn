```python
import pandas as pd
import numpy as np


data = pd.read_csv("bank.csv")
```

```python
print(data.shape)
print(data.head())
```

```python
x = data.drop("class", axis=1)
y = data['class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

```python
pred = model.predict(x_test)
data_pred = pd.DataFrame({"Actuals":y_test, "Predictions":pred})
print(data_pred.head(10))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(pred, y_test))
print(classification_report(pred, y_test))
```

```python
# вывод матицы ошибок 

import matplotlib.pyplot as plt
from sklearn import metrics

cm = metrics.confusion_matrix(y_test, pred)
# initialize confusion_matrix as a variable cm, which contains the error matrix y_test and pred
# and also the output of labels on graphics is initialized as False and True
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[False, True])
cm_display.plot()
plt.show()
```
